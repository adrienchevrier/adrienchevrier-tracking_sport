```python
import pandas as pd
```

```python
data = pd.read_csv('Activities.csv', sep=',')
```

```python
data['distance km'].apply()
```

```python
def to_km (act,di):
    if act=='En piscine':
        res = pd.to_numeric(di.replace(',',''))/1000
        return res
    else:
        return di
```

```python
data['Distance km'] = data.apply(lambda x: to_km(x["Type d\'activité"],x["Distance"]),axis=1)
```

```python
data.to_csv('ActivitiesB.csv',sep=',')
```
